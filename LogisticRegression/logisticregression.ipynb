```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("bikebuyer1.csv")
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.describe
```

```python
#check for any missing value
dataset.isnull().any()
```

```python
#filling missing values with mean, median and mode
dataset['Marital Status'].fillna(dataset['Marital Status'].mode()[0],inplace = True)
dataset['Gender'].fillna(dataset['Gender'].mode()[0],inplace = True)
dataset['Children'].fillna(dataset['Children'].median(),inplace = True)
dataset['Commute Distance'].fillna(dataset['Commute Distance'].mean(),inplace = True)
```

```python
dataset.isnull().any()
```

```python
dataset.drop(['ID','Education','Home Owner'],axis=1,inplace = True)
```

```python
dataset
```

```python
dataset['Marital Status'].unique()
```

```python
dataset['Gender'].unique()
```

```python
dataset['Occupation'].unique()
```

```python
dataset['Region'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Marital Status'] = le.fit_transform(dataset['Marital Status'])
dataset['Gender'] = le.fit_transform(dataset['Gender'])
dataset['Occupation'] = le.fit_transform(dataset['Occupation'])
dataset['Region'] = le.fit_transform(dataset['Region'])
dataset['Bike Buyer'] = le.fit_transform(dataset['Bike Buyer'])
```

```python
dataset.head(5)
```

```python
x = dataset.iloc[:,0:9].values
y = dataset.iloc[:,9].values
```

```python
x.shape
```

```python
#Using one hot encoder on Occupation and Region
from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder()
z = one.fit_transform(x[:,4:5]).toarray()
z1 =one.fit_transform(x[:,7:8]).toarray()
x = np.delete(x,[4,7],axis=1)
x = np.concatenate((z,z1,x),axis=1)
```

```python
x.shape
```

```python
#split the data to train and set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state= 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
y_pred = log.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
#find the accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
#Graph visualization
import matplotlib.pyplot as plt
plt.plot(x_test[:,14],y_pred)
plt.show()
```

```python
#Check with own values
y_pre = log.predict(sc.transform([[0,1,1,0,1,0,0,0,1,0,4000,4,0,2,28]]))
```

```python
y_pre
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
import sklearn.metrics as metrics
fpr, tpr, threshold = metrics.roc_curve(y_test,y_pred) #fpr = false positive rate, trp = true positive rate 
roc_auc = metrics.auc(fpr,tpr)
```

```python
plt.title("roc")
plt.plot(fpr,tpr,'b',label = 'Auc = %0.2f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([0,1])
plt.ylim([0,1])
plt.xlabel('tpr')
plt.ylabel('fpr')
plt.show()
```
